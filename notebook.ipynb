Shor Simulation
===


```python
import shor

import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.empty(512, dtype=np.double)
for k in range(512):
    if (k-1) % 6 == 0:
        p[k] = 1/86
    else:
        p[k] = 0
        
plt.plot(p)
plt.show()

p = shor.DFT(p, 512)
plt.plot(p)
plt.show()
```
