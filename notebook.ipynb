Shor Simulation
===


```python
import shor

import math
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.empty(512, dtype=np.double)
for k in range(512):
    if (k-1) % 6 == 0:
        p[k] = 1/86
    else:
        p[k] = 0
        
plt.plot(p)
plt.show()

p = shor.DFT(p, 512)
plt.plot(p)
plt.show()
```

```python
N = int(input("N: "))
x = int(input("x: "))
n = math.ceil(math.log2(N))
t = math.floor(2 * math.log2(N)) + 2

n_states = 2 ** t 
QReg = {"n_states": n_states,
        "prob": np.empty(n_states, dtype=np.double)}
```

```python
def phi_0(reg):
    n_states = reg["n_states"]
    prob = reg["prob"]
    
    prob[0] = 1
    for j in range(1, n_states):
        prob[j] = 0
```

```python
phi_0(QReg)
plt.plot(QReg["prob"], 'r')
plt.show()
```

```python
def phi_1(reg):
    n_states = reg["n_states"]
    prob = reg["prob"]
    
    for j in range(n_states):
        prob[j] = 1 / n_states
```

```python
phi_1(QReg)
plt.plot(QReg["prob"], 'r')
plt.show()
```

```python
def phi_2(reg):
    n_states = reg["n_states"]
    prob = reg["prob"]
    freq_powers = np.empty(N, dtype=np.double)
    for j in range(n_states):
        power = shor.power_mod(x, j, N)
        freq_powers[power] += prob[j]
    return freq_powers
```

```python
freq_powers = phi_2(QReg)
plt.plot(freq_powers, 'ro')
plt.show()
```

```python
def phi_3(reg):
    n_states = reg["n_states"]
    prob = reg["prob"]
    
    j = random.randint(0, n_states-1)
    x_b = shor.power_mod(x, j, N)
    
    cnt = 0
    for j in range(n_states):
        if x_b == shor.power_mod(x, j, N):
            cnt += 1
            
    for j in range(n_states):
        if x_b == shor.power_mod(x, j, N):
            prob[j] = 1 / cnt
        else:
            prob[j] = 0
```

```python
phi_3(QReg)
plt.plot(QReg["prob"], 'r')
plt.show()
```

```python
def phi_4(reg):
    n_states = reg["n_states"]
    reg["prob"] = shor.DFT(reg["prob"], n_states)
```
